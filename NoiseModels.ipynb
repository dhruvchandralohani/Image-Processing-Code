```python
import numpy as np
import cv2 as cv
from scipy import ndimage as nd
from matplotlib import pyplot as plt
```

```python
def displayImage(img):
  """
  Function to display image.
  Input: Image
  Output: Plot of Image.
  """
  plt.imshow(img,cmap='gray')
  plt.xticks([])
  plt.yticks([])
```

```python
img = cv.imread('Fig0427(a)(woman).tif',0)
plt.figure(figsize=(7,7))
displayImage(img)
plt.show()
```

```python
# Salt and Pepper Noise
plt.figure(figsize=(15,15))
deg_img1 = np.copy(img)

row, col = img.shape
snp_noise = np.random.choice((1, 2, 3), size=(row, col), p=[0.1, 0.1, 0.8])
deg_img1[snp_noise==1] = 0
deg_img1[snp_noise==2] = 255
plt.subplot(1, 2, 1)
displayImage(deg_img1)

filt_img1 = cv.medianBlur(deg_img1, 5)
plt.subplot(1, 2, 2)
displayImage(filt_img1)

plt.show()
```

```python
# Gaussian Noise
plt.figure(figsize=(15,15))
deg_img2 = np.copy(img)

row, col = img.shape
gauss_noise = np.random.normal(0, 50, size=(row, col))
deg_img2 = deg_img2 + gauss_noise
plt.subplot(1, 2, 1)
displayImage(deg_img2)

filt_img2 = nd.gaussian_filter(deg_img2, 3)
plt.subplot(1, 2, 2)
displayImage(filt_img2)

plt.show()
```

```python
# Uniform Noise
plt.figure(figsize=(15,15))
deg_img3 = np.copy(img)

row, col = img.shape
uniform_noise = np.random.uniform(0, 255, size=(row, col))
deg_img3 = deg_img3 + uniform_noise
plt.subplot(1, 2, 1)
displayImage(deg_img3)

plt.show()
```

```python
# Rayleigh Noise
plt.figure(figsize=(15,15))
deg_img4 = np.copy(img)

row, col = img.shape
rayleigh_noise = np.random.rayleigh(size=(row, col))
deg_img4 = deg_img3 + rayleigh_noise
plt.subplot(1, 2, 1)
displayImage(deg_img4)

plt.show()
```

```python
# Gamma(Erlang) Noise
plt.figure(figsize=(15,15))
deg_img5 = np.copy(img)

row, col = img.shape
gamma_noise = np.random.gamma(1000, size=(row, col))
deg_img5 = deg_img5 + gamma_noise
plt.subplot(1, 2, 1)
displayImage(deg_img5)

plt.show()
```

```python
# Exponential Noise
plt.figure(figsize=(15,15))
deg_img6 = np.copy(img)

row, col = img.shape
exp_noise = np.random.exponential(50, size=(row, col))
deg_img6 = deg_img6 + exp_noise
plt.subplot(1, 2, 1)
displayImage(deg_img6)

plt.show()
```
