```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def displayImage(img):
  """
  Function to display image.
  Input: Image
  Output: Plot of Image.
  """
  plt.imshow(img,cmap='gray')
  plt.xticks([])
  plt.yticks([])
```

```python
img = cv.imread('NZ.jpg', 0)
plt.figure(figsize=(20, 10))

plt.subplot(1, 4, 1)
displayImage(img)

plt.subplot(1, 4, 2)
plt.hist(img)

plt.subplot(1, 4, 3)
hist1, bins1 = np.histogram(img.ravel(), 256, [0, 256])
plt.plot(hist1)

plt.subplot(1, 4, 4)
hist2 = cv.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist2)

plt.show()
```

```python
# Histogram Equalization
equ = cv.equalizeHist(img)

plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
displayImage(img)

plt.subplot(1, 2, 2)
displayImage(equ)
```
