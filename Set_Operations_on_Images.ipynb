```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def createBinaryImage():
  """
  Function to generate a binary image.
  Input: Dimension of image.
  Output: Plot of the image.
  """
  dim = [int(item) for item in input("Enter Dimension : ").split()]
  binaryImage = np.zeros([dim[0], dim[1]])
  Xcord = np.random.randint(0, dim[0], dim[0]*10)
  Ycord = np.random.randint(0, dim[1], dim[1]*10)
  binaryImage[Xcord, Ycord] = 1
  plt.imshow(binaryImage, cmap = 'gray')
  plt.show()
  cv.imwrite('C:\College\Digital Image Processing\Images\TestImage.tif', binaryImage)

createBinaryImage()
```

```python
def displayImage(img):
  """
  Function to display image.
  Input: Image
  Output: Plot of Image.
  """
  plt.imshow(img,cmap='gray')
  plt.xticks([])
  plt.yticks([])
```

```python
def Normalize(img):
  """
  Function to perform Min-Max feature scaling.
  """
  return (img - np.min(img))/(np.max(img) - np.min(img))
```

```python
def BinaryImage(img):
  """
  Function to create image having intensities 0 and 1.
  Input: Image
  Output: Binary Image.
  """
  img = Normalize(img)
  sza = np.shape(img)
  for i in range(sza[0]):
    for j in range(sza[1]):
      if img[i,j] >= 0.5:
        img[i,j] = 1
      else:
        img[i,j] = 0
  return img

img = cv.imread('lenna.tif',0)
plt.figure(figsize=(20,20))
plt.subplot(1, 2, 1)
displayImage(img)

ImgBin = BinaryImage(img)
plt.subplot(1, 2, 2)
displayImage(ImgBin)

plt.show()
```

```python
def union(img):
  """
  Function to perform union operation on two binary images. Second image is mirror image of first image.
  Input: Image.
  Output: Original image and it's mirror image in binary, union of these two images.
  """
  img1 = img
  img2 = np.flip(img1, axis = 1)
  img1 = BinaryImage(img1)
  img2 = BinaryImage(img2)
  sze = np.shape(img)
  img3 = np.zeros(sze)
  for i in range(sze[0]):
    for j in range(sze[1]):
      if img1[i,j] == 1 or img2[i,j] == 1:
        img3[i,j] = 1
  return img1, img2, img3

img = cv.imread('lenna.tif',0)
img1, img2, img3 = union(img)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
displayImage(img)     # Original Image.(img)
plt.subplot(1, 4, 2)
displayImage(img1)    # Original Image in binary.(img1)
plt.subplot(1, 4, 3)
displayImage(img2)    # Mirror Image of img1.(img2)
plt.subplot(1, 4, 4)
displayImage(img3)    # Union of img1 and img2.(img3)
plt.show()
```

```python
def intersection(img):
  """
  Function to perform intersection operation on two binary images.Second image is mirror image of first image.
  Input: Image.
  Output: Original image and it's mirror image in binary, intersection of these two images.
  """
  img1 = img
  img2 = np.flip(img1, axis = 1)
  img1 = BinaryImage(img1)
  img2 = BinaryImage(img2)
  sze = np.shape(img)
  img3 = np.zeros(sze)
  for i in range(sze[0]):
    for j in range(sze[1]):
      if img1[i,j] == 1 and img2[i,j] == 1:
        img3[i,j] = 1
  return img1, img2, img3

img = cv.imread('lenna.tif',0)
img1, img2, img3 = intersection(img)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
displayImage(img)     # Original Image.(img)
plt.subplot(1, 4, 2)
displayImage(img1)    # Original Image in binary.(img1)
plt.subplot(1, 4, 3)
displayImage(img2)    # Mirror Image of img1.(img2)
plt.subplot(1, 4, 4)
displayImage(img3)    # Intersection of img1 and img2.(img3)
plt.show()
```

```python
def complement(img):
  """
  Function to perform complement operation on a binary image.
  Input: Image.
  Output: Original image in bianry and it's complement image.
  """
  img1 = BinaryImage(img)
  sze = np.shape(img)
  img3 = np.zeros(sze)
  for i in range(sze[0]):
    for j in range(sze[1]):
      if img1[i,j] == 1 :
        img3[i,j] = 0
      else:
        img3[i,j] = 1
  return img1, img3

img = cv.imread('lenna.tif',0)
img1, img3 = complement(img)

plt.figure(figsize=(20,20))
plt.subplot(1, 3, 1)
displayImage(img)     # Original Image.(img)
plt.subplot(1, 3, 2)
displayImage(img1)    # Original Image in binary.(img1)
plt.subplot(1, 3, 3)
displayImage(img3)    # Complement Image of img1.(img3)
plt.show()
```

```python
def xor(img):
  """
  Function to perform XOR operation on two binary images. Second image is mirror image of first image.
  Input: Image.
  Output: Original image and it's mirror image in binary, union of these two images.
  """
  img1 = img
  img2 = np.flip(img1, axis = 1)
  img1 = BinaryImage(img1)
  img2 = BinaryImage(img2)
  sze = np.shape(img)
  img3 = np.zeros(sze)
  for i in range(sze[0]):
    for j in range(sze[1]):
      if img1[i,j] == img2[i,j]:
        img3[i,j] = 0
      else:
        img3[i,j] = 1
  return img1, img2, img3

img = cv.imread('lenna.tif',0)
img1, img2, img3 = union(img)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
displayImage(img)     # Original Image.(img)
plt.subplot(1, 4, 2)
displayImage(img1)    # Original Image in binary.(img1)
plt.subplot(1, 4, 3)
displayImage(img2)    # Mirror Image of img1.(img2)
plt.subplot(1, 4, 4)
displayImage(img3)    # XOR of img1 and img2.(img3)
plt.show()
```
