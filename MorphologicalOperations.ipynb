```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
def displayImage(img):
  """
  Function to display image.
  Input: Image
  Output: Plot of Image.
  """
  plt.imshow(img,cmap='gray')
  plt.xticks([])
  plt.yticks([])
```

```python
# Erosion
plt.figure(figsize=(15, 15))
img1 = cv.imread('Fig0905(a)(wirebond-mask).tif', 0)
plt.subplot(1, 2, 1)
displayImage(img1)

kernel1 = np.ones((3,3),np.uint8)
eroded_img = cv.erode(img1,kernel1,iterations = 1)
plt.subplot(1, 2, 2)
displayImage(eroded_img)

plt.show()
```

```python
# Dilation
plt.figure(figsize=(15, 15))
img2 = img1 = cv.imread('Fig0907(a)(text_gaps_1_and_2_pixels).tif', 0)
plt.subplot(1, 2, 1)
displayImage(img2)

kernel2 = np.ones((3,3),np.uint8)
dilated_img = cv.dilate(img2,kernel2,iterations = 1)
plt.subplot(1, 2, 2)
displayImage(dilated_img)

plt.show()
```

```python
# Opening
plt.figure(figsize=(15, 15))
img3 = cv.imread('Fig0911(a)(noisy_fingerprint).tif', 0)
plt.subplot(1, 2, 1)
displayImage(img3)

kernel3 = np.ones((3,3),np.uint8)
opened_img = cv.morphologyEx(img3, cv.MORPH_OPEN, kernel3)
plt.subplot(1, 2, 2)
displayImage(opened_img)

plt.show()
```

```python
# Closing
plt.figure(figsize=(15, 15))
img4 = cv.imread('FigP0936(bubbles_on_black_background).tif', 0)
plt.subplot(1, 2, 1)
displayImage(img4)

kernel4 = np.ones((7,7),np.uint8)
closed_img = cv.morphologyEx(img4, cv.MORPH_CLOSE, kernel4)
plt.subplot(1, 2, 2)
displayImage(closed_img)

plt.show()
```

```python
# Thinning
plt.figure(figsize=(15, 15))
img5 = cv.imread('FigP0905(U).tif', 0)
plt.subplot(1, 2, 1)
displayImage(img5)

thinned_img = cv.ximgproc.thinning(img5, thinned_img, cv.ximgproc.THINNING_ZHANGSUEN)
plt.subplot(1, 2, 2)
displayImage(thinned_img)

plt.show()
```
