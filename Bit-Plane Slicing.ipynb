```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
def displayImage(img):
  """
  Function to display image.
  Input: Image
  Output: Plot of Image.
  """
  plt.imshow(img,cmap='gray')
  plt.xticks([])
  plt.yticks([])
```

```python
def Normalize(img):
  """
  Function to perform Min-Max feature scaling.
  """
  return (img - np.min(img))/(np.max(img) - np.min(img))
```

```python
img = cv.imread('dollar.tif',0)
plt.figure(figsize=(20,20))
displayImage(img)
plt.show()

plt.figure(figsize=(30,30))

#Bit-plane 1.
img1 = img & 1
plt.subplot(1, 4, 1)
displayImage(img1)

#Bit-plane 2.
img2 = img & 2
plt.subplot(1, 4, 2)
displayImage(img2)

#Bit-plane 3.
img3 = img & 4
plt.subplot(1, 4, 3)
displayImage(img3)

#Bit-plane 4.
img4 = img & 8
plt.subplot(1, 4, 4)
displayImage(img4)

plt.show()
plt.figure(figsize=(30,30))

#Bit-plane 5.
img5 = img & 16
plt.subplot(1, 4, 1)
displayImage(img5)

#Bit-plane 6.
img6 = img & 32
plt.subplot(1, 4, 2)
displayImage(img6)

#Bit-plane 7.
img7 = img & 64
plt.subplot(1, 4, 3)
displayImage(img7)

#Bit-plane 8.
img8 = img & 128
plt.subplot(1, 4, 4)
displayImage(img8)

plt.show()
```
